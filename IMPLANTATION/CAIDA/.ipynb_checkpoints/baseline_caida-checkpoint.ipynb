###Choix de la base de données sur laquelle travailler 

```python
#Définition des paths

#le dataset doit se trouver dans le répertoire IMPLANTATION/CAIDA/data_GCN/ du répertoire github
def load_dataset(nom_dataset):
  groupRepoPath = "groupe/"
  pathData = os.path.join(groupRepoPath,"IMPLANTATION/CAIDA/data_GCN/")
  #path vers le fichier qui contient les données 
  pathDataset = os.path.join(pathData,nom_dataset)
  #on prend la dernière version du repositoire
  if os.path.exists(groupRepoPath):
    shutil.rmtree(groupRepoPath)
  Repo.clone_from("https://github.com/Viperine2022/projet_long_GCN_internet.git",groupRepoPath)
  #on chagge les données dans un dataframe depuis un fichier en format CSV
  dataset = pd.read_csv(pathDataset)
  return dataset
  
```

##premier decoupage en general train et test set 

il faut vérifier qu'en prénant différent test sets les résultats ne sont pas très différents si c'est le cas --> utiliser methode K-Folds / version shuffled (utile surtout pour petits datasets)

```python
#Affichage des proportions des classes dans le dataset originel 
print(supervised['label'].value_counts())

#ligne pour transformer les classes CAIDA en entiers si cela n'a pas été fait en pretraitement 
supervised['label'] = supervised['label'].map({'Content':0,'Transit/Access':1,'Enterprise':2})


#on verifie que les labels ont le format correct pour les passer au modèle 
print(supervised['label'])

#choix d'un entier aléatoire pour le decoupage qu'on gardera aprés si on veut rejouer des verifiactions sur une partition en particulier 
etat_rand = random.randint(0, 101)

#partitionnement
train_set, test_set = train_test_split(supervised, test_size=0.2, random_state=etat_rand)
```

##visualisation classes

```python
#Visualisation de la decompensation de classes 
nbContent = len(train_set[train_set.label == 0])
nbTransit = len(train_set[train_set.label == 1])
nbEntreprise = len(train_set[train_set.label == 2])
per_class = pd.Series([nbContent,nbTransit,nbEntreprise])

#affichage de la proportion de classes au sein de l'ensemble 
print(per_class)

```

```python
#Affichage de la pie chart avec les proportions des classes 
plt.figure(figsize=(20,10))
my_circle=plt.Circle( (0,0), 0.7, color='white')
plt.pie(per_class, labels=['Content providers ', 'Internet Providers', 'Enterprise'],autopct='%1.1f%%')
p=plt.gcf()
p.gca().add_artist(my_circle)
plt.show()
```

##oversampling
On genere des echantillons dans les classes déficitaires pour avoir la même quantité de données dans toutes les classes 

```python
#géneration des random states pour le prelevement du sampling 
randStates = np.random.randint(0,101,3)

#Choix du nombre d'echantillons entre la quantité de representants de la première et la deuxième classe 
NombreSamples = 10000

df_contents =train_set[train_set['label']==0]
df_entreprise =train_set[train_set['label']==1]
df_transit =train_set[train_set['label']==2].sample(n=NombreSamples,replace=True,random_state=randStates[0])

#comme on a moins de 20.000 elements dans affectés aux classes Entreprise et Content providers on va generer des samples additionels pour ces classes 
df_contents_upsample=resample(df_contents,replace=True,n_samples=NombreSamples,random_state=randStates[1])
df_entreprise_upsample=resample(df_entreprise,replace=True,n_samples=NombreSamples,random_state=randStates[2])

train_set=pd.concat([df_contents_upsample,df_transit,df_entreprise_upsample])

#Vérification sur la taille des données 
print(train_set.shape)

#Affichage de la nouvelle proportion des classes
nbContent = len(train_set[train_set.label == 0])
nbTransit = len(train_set[train_set.label == 1])
nbEntreprise = len(train_set[train_set.label == 2])
per_class = pd.Series([nbContent,nbTransit,nbEntreprise])
plt.figure(figsize=(20,10))
my_circle=plt.Circle( (0,0), 0.7, color='white')
plt.pie(per_class, labels=['Content providers ', 'Internet Providers', 'Enterprise'],autopct='%1.1f%%')
p=plt.gcf()
p.gca().add_artist(my_circle)
plt.show()
```

##downsampling
On restraint le nombre d'échatillons de chaque classe à la classe ayant le plus petit nombre d'échantillons 

```python
#géneration des random states pour le prelevement du sampling 
randStates = np.random.randint(0,101,3)

#downsampling: on prend les 4198 premiers valeurs de chaque classe 
nbSamples = min(train_set['label'].value_counts())


df_contents =train_set[train_set['label']==0].sample(n=nbSamples,random_state=randStates[0],replace=True)
df_entreprise =train_set[train_set['label']==1].sample(n=nbSamples,random_state=randStates[1],replace=True)
df_transit =train_set[train_set['label']==2].sample(n=nbSamples,random_state=randStates[2],replace=True)
train_set=pd.concat([df_transit,df_contents,df_entreprise])
```

##separation données train

Selection de l'ensemble d'attributs à prendre en compte pour l'entraînement

```python
#Configurations d'attributs pour l'entraînement 
atributsLred = ['page_rank_not_directed', 'degree_centrality', 'peering_links', 'customer_links', 'provider_links']
atributsL = ['page_rank_not_directed', 'page_rank_directed','page_rank_directed_inverse', 'degree_centrality', 'peering_links',
             'customer_links', 'provider_links']
atributsPage = ['page_rank_not_directed', 'page_rank_directed','page_rank_directed_inverse', 'degree_centrality']
atributsRatio = ['page_rank_not_directed','page_rank_directed', 'page_rank_directed_inverse',
                 'degree_normalized','ratio_peering', 'ratio_customer','ratio_provider']
atributsJusteRatio = ['degree_normalized','ratio_peering', 'ratio_customer','ratio_provider']    
uniqueatr = ['degree_centrality'] 

#Ligne pour choisir les attributs 
atributs = atributsRatio 


train_set_final, validation_set = train_test_split(train_set, test_size=0.2, random_state=etat_rand)

print("le nombre d'atributs utilisés dans l'execution est de: ",len(atributs))
train_set_final.head()
x_train = train_set_final[atributs]
y_train = train_set_final['label']

#définition des ins et outs pour la validation
x_val = validation_set[atributs]
y_val = validation_set['label']

#définition des ins et outs pour le test
x_test = test_set[atributs]
y_test = test_set['label']

#affichage des premiers valeurs des attributs 
print("dataframe des attributs ",x_train.head())
#affichage des premiers valeurs des labels 
print("dataframe des labels ",y_train.head())

```

# MODEL TRAINING 

##Multi Layer Perceptron with dense layers 

l2=1e-4  kernel_regularizer=regularizers.l1_l2(l1=1e-5) 
kernel_regularizer='l1'

Information utile pour la conception: 



*   intermediate layers too small => information bottleneck 
*   petit dataset : peu de layers | severe overfitting  & K-fold validation est utile
* dense layers + relu activation --> common . But cuts the negative values
* diminuer la taille du modele pour l'étape underfiting (commencer petit puis augmenter puis reduire autant que possible)
* ajouter des couches de dropout souvent entre 0.2 et 0.5


 


```python
#definition du modèle 
model = keras.Sequential(
    [Dense(input_dim=len(atributs),units =50,activation='tanh'),
     Dropout(0.2),
     Dense(50,activation='tanh'),
     Dropout(0.2),
     Dense(50,activation='tanh'),
     Dense(3,activation='softmax')
])
```

## Compilation du modèle 

configuration de l'étape d'entraînement où sont selectionés l'optimizer, les metrics et la fonction de perte

```python
opt = tf.keras.optimizers.Adam()
opt.learning_rate = 1e-3
model.compile(loss='sparse_categorical_crossentropy',metrics=['accuracy'],optimizer=opt)
```

## Lancement de l'entrînement 

* configuration du nombre d'épochs, du batch size et de l'input data à utiliser

* validation_data est un argument de fit qui nous permet d'introduire les données de validation

* le fit retourne un dictionnaire avec out.history qui contient les clés suivantes servant à faire des affichages : u"accuracy", u"loss", u"val_accuracy", u"val_loss"

* Le debugging d'un modèle peut se faire par l'utilisation de la methode model.sumary() à utiliser après le build du modéle

```python
epochs_ = 200
output = model.fit(x=x_train, y=y_train, batch_size=1000, epochs=epochs_,validation_data = [x_val,y_val])
history_dict = output.history

#affichage des résultats de l'entraînement 
print(history_dict)

#affichage des figures : 
plt.plot(range(epochs_), history_dict["accuracy"], "bo", label="Training acc")
plt.plot(range(epochs_), history_dict["val_accuracy"], "b+", label="Validation acc")
plt.title("Training and validation accuracy")
plt.xlabel("Epochs")
plt.ylabel("Accuracy")
plt.legend()
plt.show()
#netoiages de la figure avant d'afficher la suivante 
plt.clf() 
plt.plot(range(epochs_), history_dict["loss"], "ro", label="Training loss")
plt.plot(range(epochs_), history_dict["val_loss"], "r+", label="Validation loss")
plt.xlabel("Epochs")
plt.ylabel("Loss")
plt.legend()
plt.show()
```

## Évaluation sur l'ensemble de tests 
Nous validons le modèle sur l'ensemble de test


Résultat de la fonction evaluate : 


> [test loss, test accuracy]



```python
eval = model.evaluate(x_test,y_test)

print(eval)
```

## Vérification de la puissance d'inférence 
capacité du modèle à produire des nouvelles valeurs. En particulier on peut regarder comment se répartissent les prédictions correctes parmi les classes disponibles à l'aide de la matrice de confussion et du classification report qui affiche le recall et la précision d'un prédiction par rapport à chaque classe 

```python
#verification de la puissance d'inference (test du modèle pour prédire des nvx données) 
predictions = model.predict(x_test)

#déduction de la classe dont la proba est majoritaire  
predictedClass = np.argmax(predictions,axis=1)
#on verifie que la classe des predictions a la bonne forme
print(predictedClass)
#on comptabilise les resultats par classe
uniq,freq = np.unique(predictedClass,return_counts=True)
#affichage des resultats 
print("Valeurs uniques predicted: ",uniq)
print("Frequence valeurs predicted: ",freq)

#resample les classes pour avoir la meme quantité 
y_test_np = y_test.to_numpy()
trueClass = y_test_np 
uniq,freq = np.unique(trueClass,return_counts=True)
#proportion des labels dans la classe de test
print("Valeurs uniques dans les test: ",uniq)
print("Frequence valeurs dans les test: ",freq)
```

```python
#affichage de la matrice de confusion 
print(trueClass)
classe = ["Content Provider","Transit Access","Entreprise"]
N_CLASSES = 3
print('Matrice de confusion')
conf=confusion_matrix(trueClass,predictedClass)
sns.heatmap(data=conf,fmt='.0f',xticklabels=classe,yticklabels=classe,annot=True)
# score de performance
print('Accuracy sur base de test :',accuracy_score(trueClass,predictedClass))
```

```python
sns.heatmap(conf/(np.sum(conf,axis=1)[:,None]), annot=True, 
            fmt='.2%', cmap='Blues',xticklabels=classe,yticklabels=classe)
```

```python
#affichage du report de classification. Ici on regarde principalement le recall 
cr = classification_report(trueClass,predictedClass,target_names=classe)
print(cr)
```
